```python
# dropping the null values from the data
for col in df.columns:
    if(df[col].isnull().sum()/len(df.index)*100) > 50:
        df.drop(col,axis=1,inplace=True)
```

```python
print("Country with the most attacks:",df['country_txt'].value_counts().idxmax())
# as first entry is 'unknown' using index as 1
print("City with the most attacks:",df['city'].value_counts().index[1]) 
print("Region with the most attacks:",df['region_txt'].value_counts().idxmax())
print("Year with the most attacks:",df['iyear'].value_counts().idxmax())
print("Month with the most attacks:",df['imonth'].value_counts().idxmax())
print("Group with the most attacks:",df['gname'].value_counts().index[1])
print("Most Attack Types:",df['attacktype1_txt'].value_counts().idxmax())
```

1. Checking the number of attacks in a year

```python
plt.figure(figsize=(15,5))
sns.countplot(df['iyear'])
plt.xticks(rotation = 45)
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks each year', fontsize=8)
plt.title('Attacks In Years',fontsize=12)
plt.show()
```

We can analyze that in recent years the attack was increasing. The most number of attacks took place in year 2014.

2. Terrorist Activities by Region in each year

```python
pd.crosstab(df.iyear, df.region_txt).plot(kind='area',figsize=(15,6))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(df['region_txt'])
plt.xticks(rotation = 45)
plt.xlabel('Region',fontsize=3)
plt.ylabel('Number of Attacks', fontsize=8)
plt.title('Region Vs Attack',fontsize=12)
plt.show()
```

The highest number of attack happening in middle East & North Africa and South Asia

```python
7. Terror Attacks by Taliban over the years.
```

```python
df_tali = df[df.gname == 'Taliban']
```

```python

plt.figure(figsize = (13,7))
sns.barplot(df_tali['iyear'].value_counts().index,df_tali['iyear'].value_counts().values, palette = 'viridis')
plt.title('Terror Attack over the years by Taliban',fontsize=15)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Number of Attacks',fontsize=15)
plt.xticks(rotation= 90)
```

8. The number of deaths(killed) occured by Tabliban over the years and in cities

```python
plt.figure(figsize=(13,7))
df_tali.groupby(['iyear'])['nkill'].sum().plot(kind='bar',colormap='RdBu')
plt.title('People Killed by Taliban over the Years',fontsize=15)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Number of people killed',fontsize=15)
plt.xticks(rotation=90)
plt.grid()
plt.show()
```
